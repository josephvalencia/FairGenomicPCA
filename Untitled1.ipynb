```python
train_npz = np.load("adult.data.npz")
test_npz = np.load("adult.test.npz")
```

```python
train_data = train_npz['data']
test_data = test_npz['data']
```

```python
from sklearn.preprocessing import scale

train_data = scale(train_data)
```

```python
train_data[:,1]
```

```python
from sklearn.decomposition import PCA

# plot sklearn PCA
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection="3d")

pca = PCA(n_components=3)
pca.fit(train_data)
p_components = pca.components_.T

projected = train_data @ p_components
reconstruction_loss = loss(train_data,p_components)
print("sklearn reconstruction",reconstruction_loss)

a,b,c = np.split(projected,3,1)
ax.scatter3D(-c,b,-a,s=5)
plt.show()

# plot my PCA
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection="3d")
eigvals,eigvecs = run_PCA(train_data,3)

transformed = train_data @ eigvecs
reconstruction_loss = loss(train_data,eigvecs)
print("my reconstruction loss",reconstruction_loss)

x,y,z = np.split(transformed,3,1)
ax.scatter3D(x,-y,z,s=5)
plt.show()
```
